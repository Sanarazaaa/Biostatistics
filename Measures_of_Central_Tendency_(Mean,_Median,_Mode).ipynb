```python
import pandas as pd
df = pd.read_csv("Cell_Observations.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
mean_value = df ['X_position_of_cell'].mean()
median_value = df['X_position_of_cell'].median()
mode_value = df ['X_position_of_cell'].mode()[0]

print("Mean:", mean_value)
print('Median:', median_value)
print ('Mode:', mode_value)
```

*We prefer the median over the mean when data contains outliers or is skewed, as the median is less sensitive to extreme values and provides a more robust representation of the central tendency in such cases. Here the mean is a bit higher than the the median. This means that data might be right skewed.*

Now, we can check for skeweness.

1. If Skewness > 0, the data is right-skewed (outliers on the right)
2. If Skewness < 0, the data is left-skewed
3. If Skewness â‰ˆ 0, the data is roughly symmetric.

```python
print("Skewness:", df['X_position_of_cell'].skew())
```

```python
# We can also check through boxplot. If there are individual dots outside the whiskers, then that means that it has an outlier.

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.boxplot(y=df['X_position_of_cell'])
plt.title("Boxplot of X_position_of_cell")
plt.show()
```

**The boxplot for X_position_of_cell shows:**

* The median (middle line inside the box) is around 49, which matches calculated median (49.16).

* The box is symmetric, meaning the data distribution is quite balanced.

* No significant outliers (no individual dots outside the whiskers).

**What This Means:**

âœ… Mean (49.65) and median (49.16) are close â†’ No major skewness

âœ… Mode (16.84) is lower â†’ A common value appears frequently but doesn't represent the data well

âœ… No extreme outliers â†’ The mean is a reliable measure of central tendency
